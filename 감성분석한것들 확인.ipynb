```python
import pandas as pd
import numpy as np
df=pd.read_csv('최종감성분석.csv')
df2=pd.read_csv('최종 데이터(for추천).csv')
```

# 리뷰 없음이 포함된 데이터

```python
df_a=df[df['리뷰내용_1'].str.contains('리뷰없음')]
```

```python
df_a
```

```python
df_a['최종긍정확률']=df_a['긍정확률_0']
df_a['최종부정확률']=df_a['부정확률_0']
```

```python
df_a.head()
```

```python
df_a['리뷰']=df_a['리뷰내용_0']
```

```python
df_a.drop(['긍정확률_0','부정확률_0','긍정확률_1','부정확률_1'],axis=1,inplace=True)
```

```python
df_a
```

# 정상적인 데이터

```python
df_b=df[~df['리뷰내용_1'].str.contains('리뷰없음')]
df_b.head()
```

```python
df_b['최종긍정확률']=(df_b['긍정확률_0']+df_b['긍정확률_1'])/2
df_b['최종부정확률']=(df_b['부정확률_0']+df_b['부정확률_1'])/2
```

```python
df_b['리뷰']=df_b['리뷰내용_0'] + df_b['리뷰내용_1']
```

```python
df_b.head()
```

```python
df_b.drop(['긍정확률_0','부정확률_0','긍정확률_1','부정확률_1'],axis=1,inplace=True)
```

```python
df_b
```

```python
df=pd.concat([df_b,df_a])
df
```

```python
df.info()
```

```python
df=df[['관광지','리뷰','최종긍정확률','최종부정확률']]
```

```python
df.head()
```

```python
df_group=df.groupby(['관광지'],as_index=False).mean()
```

```python
df_group
```

```python
df_group.rename(columns={'관광지':'여행지'},inplace=True)
```
